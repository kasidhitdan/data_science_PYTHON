```python
df = df.reset_index()
```

```python
#Create a smaller DataFrame as a subset of the original df
small_df =df[['title','release_date','budget','revenue','runtime','genres']]
#Display the data in tabular format of small_df
small_df.head(15)
```

```python
#Check the data types of the dataframe
small_df.info()
```

```python
#Import the numpy library 
import numpy as np
#Create a function to check values and convert them to float
def to_float(x) :
    try:
        x = float(x)
    except:
        x = np.nan
    return x
#Use the apply method to check and convert every value in the budget column using the to_float() function
small_df['budget'] = small_df['budget'].apply(to_float)
#Convert the data type of the budget column to float
small_df['budget'] = small_df['budget'].astype('float')
#Check data type from column
small_df.info()
```

```python
#Convert release_date to pandas datetime format first
small_df['release_date'] =pd.to_datetime(small_df['release_date'], errors='coerce')
#Extract the year data from release_date to create a new column named 'year'
small_df['year'] = small_df['release_date'].apply(lambda x:
str(x).split('-')[0] if x != np.nan else np.nan)
#Display the data in tabular format of small_df.
small_df.head()
```

```python
#Arrange the data of movies by using the 'year' column
small_df = small_df.sort_values('year')
#Display the data in tabular format of small_df
small_df.head()
```

```python
#Select movies that have generated revenue greater than 1 billion dollars and have a budget less than 150 million dollars
my_movie_list = small_df[(small_df['revenue'] > 1e9) &
(small_df['budget'] < 1.5e8)]
my_movie_list
```

```python
#Check the data type of the data
print(type(small_df['runtime']))
#Assign the runtime variable to a Series object
runtime = small_df['runtime']
#Print the maximum value of the runtime of movies
print(runtime.max())
#Print the minimum value of the runtime of movies
print(runtime.min())
```

```python
#Assign the budget variable to a Series object
budget = small_df['budget']
#Print the mean value of the budget of movies
print(budget.mean())
#Print the median value of the budget of movies
print(budget.median())
```
